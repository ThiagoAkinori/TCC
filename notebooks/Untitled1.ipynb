```python
import pandas as pd
import heapq
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
reader = pd.read_csv('/home/thiagosato/Documentos/TCC/codes/teste.csv')
```

```python
reader
```

```python
neighbours = np.argpartition(reader.values, 4)[:,:4]
neighbours_df = pd.DataFrame(neighbours)
```

```python
neighbours_df
```

```python
matrix = list()
for index, row in neighbours_df.iterrows():
    line = list()
    row_list = row.to_list()
    for i in range(10):
        if i in row_list:
            line.append(1)
        else: 
            line.append(0)
    matrix.append(line) 
adj_matrix = np.array(matrix)
```

```python
graph = nx.from_numpy_matrix(adj_matrix)
nx.draw(graph, with_labels=True, font_weight='bold')
plt.savefig('self_made.png')
```
